## Visualization

```python
import matplotlib.pyplot as plt
from matplotlib  import cm
import numpy as np

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
ax.set_title("Clustering",fontsize=14)
ax.set_xlabel("X",fontsize=12)
ax.set_ylabel("Y",fontsize=12)
ax.grid(True,linestyle='-',color='0.75')

# points coordinates
x = [15000, 20000, 17000, 2000, 5000, 100000, 150000, 160000, 145000]
y=[56000, 45000, 40000, 3000, 10000, 200000, 140000, 210000, 180000]
cluster = [1, 1, 1, 2, 2, 3, 3, 3, 3]
ax.scatter(x,y,s=40,c=cluster, marker = 'o', cmap = cm.jet )

centroid_x = [20000, 3000, 150000]
centroid_y = [50000, 5000, 190000]
ax.scatter(centroid_x,centroid_y,s=60, marker = 'x', cmap = cm.jet )

plt.show()
```
